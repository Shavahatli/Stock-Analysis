```python
import numpy as np
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#PG['simple_return'] = (PG['Adj Close'] / PG['Adj Close'].shift(1)) - 1
```

```python
dow = ['AXP', 'AAPL', 'GS', 'HD', 'KO', 'MCD', 'NKE', 'PG', 'CRM', 'VZ']
dow
```

```python
df_dow = pd.DataFrame()
for i in dow:
    stocks = wb.DataReader(i, data_source = 'yahoo', start = '2022-07-18', end = '2022-07-31')
    stocks['Company'] = i
    df_dow=pd.concat([df_dow,stocks])
    
df_dow.reset_index(inplace=True)
pd.set_option('display.max_rows', 100)
df_dow
```

```python
dates_x = df_dow['Date'].dt.strftime('%m-%d-%Y').unique().tolist()
dates_x
```

```python
x_indexes = np.arange(len(dates_x))
x_indexes
```

```python
company_colors = {'AXP':'red', 'AAPL':'green', 'GS':'blue', 'HD':'yellow','KO':'orange', 'MCD':'purple','NKE':'pink','PG':'brown','CRM':'cyan', 'VZ': 'black'}
company_names = {'AXP': 'American Express Co', 'AAPL':'Apple Inc', 'GS':'Goldman Sachs Group Inc', 'HD':'Home Depot Inc', 'KO':'Coca-Cola Co', 'MCD':'McDonald’s Corp', 'NKE':'Nike Inc', 'PG':'Procter & Gamble Co', 'CRM':'Salesforce Inc', 'VZ':'Verizon Communications Inc'}
for company in df_dow['Company'].unique().tolist():
    print(company)
    plt.plot(x_indexes, df_dow.loc[df_dow.Company == company, 'Open'], company_colors[company], label=company_names[company])
plt.xlabel('Dates')
plt.ylabel('Stock Prices')
plt.title('Daily Stock Prices in July by Companies')
plt.xticks(ticks = x_indexes, labels = dates_x, rotation=45)
plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left")
plt.show()
```
